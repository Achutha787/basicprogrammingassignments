```python
#1)
def is_symmetrical(number):
    ''' a function that takes a number as an argument and returns True or False depending on whether the number is 
    symmetrical or not. A number is symmetrical when it is the same as its reverse.'''
    if str(number) == str(number)[::-1]:
        print(f'{number}= {True}')
    else:
        print(f'{number}={False}')

is_symmetrical(7227)
is_symmetrical(12567)
is_symmetrical(44444444)
is_symmetrical(9939)
is_symmetrical(1112111)
```

```python
#2)
def multiply_nums(instrings):
    '''function for return the product of the numbers.'''
    outstrings= instrings.replace(' ','').split(',')
    outnum = 1
    try:
        for i in outstrings:
            outnum *= int(i)
    except(Exception) as e:
        print("input string only",e)
    print(f'{instrings} ➞ {outnum}')

multiply_nums("2, 3")
multiply_nums("1, 2, 3, 4")
multiply_nums("54, 75, 453, 0")
multiply_nums("10, -2")
```

```python
#3)
def square_digits(num):
    '''Create a function that squares every digit of a number.'''
    outlist = [str(int(i)**2) for i in str(num)]
    outnum=''.join(outlist)
    print(f'{num} ➞ {int(outnum)}')
square_digits(9119)
square_digits(2483)
square_digits(3212)
```

```python
#4)
def setify(numlist):
    '''a function that sorts a list and removes all duplicate items from it.'''
    result= sorted(set(numlist))
    print(f'{numlist}={result}')
setify([1, 3, 3, 5, 5]) 
setify([4, 4, 4, 4])
setify([5, 7, 8, 9, 10, 15])
setify([3, 3, 3, 2, 1])
```

```python
#5)
def mean(numbers):
    '''a function that returns the mean of all digits.'''
    digi = []
    try:
        for i in str(numbers):
            digi.append(int(i))
        result= sum(digi)/len(str(numbers))
    except(Exception) as e:
        print("pls enter integers only",e)
    print(f'{numbers} = {result}')

mean(42)
mean(12345)
mean(666)
```
