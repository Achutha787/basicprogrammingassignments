```python
#1)
def amplify(a):
    '''a function that takes an integer and returns a list from 1 to the given number, where
     1.If the number can be divided evenly by 4, amplify it by 10 (i.e. return 10 times the number).
     2. If the number cannot be divided evenly by 4, simply return the number.'''
    l1 = []
    try:
        for j in range(1,a+1):
            if j%4 == 0:
                l1.append(j*10)
            else:
                l1.append(j)
    except(Exception) as e:
        print("pls enter a integer",e)
    print(f'{a}= {l1}')
    return l1
amplify(4)
amplify(3)
amplify(25)
```

```python
#2)
def unique(list100):
    list200 = ''
    try:
        for j in set(list100):
            if list100.count(j)== 1:
                list200= j
    except(Exception) as e:
        print("pls input a list of integers",e)
    print(f'{list100} ➞ {list200}')

unique([3, 3, 3, 7, 3, 3])
unique([0, 0, 0.77, 0, 0])
unique([0, 1, 1, 1, 1, 1, 1, 1])
unique(99)
```

```python
#3)function for calculating the are and perimteter of  CIRCLE
import math
class circle:
    def __init__(self,r):
        self.r = r
    def getarea(self):
        print(f'area ➞ {round(math.pi*self.r*self.r)}')
    def getPerimeter(self):
        print(f'Perimeter ➞ {round(2*math.pi*self.r)}')
circy=circle(11)
circy.getarea()
circy = circle(4.44)
circy.getPerimeter()
```

```python
#4)
def sort_by_length(list_string):
    ''' function that takes a list of strings and return a list, sorted from shortest to longest '''
    print(sorted(list_string,key = len))
sort_by_length(["Google", "Apple", "Microsoft"])
sort_by_length(["Leonardo", "Michelangelo", "Raphael", "Donatello"])
sort_by_length(["Turing", "Einstein", "Jung"])    
```

```python
#5)
def is_triplet(a,b,c):
    '''a function that validates whether three given integers form a Pythagorean triplet.
    The sum of the squares of the two smallest integers must equal the square of the largest number to be validated.'''
    if ((a**2+b**2) == (c**2)):
        print(f'{a,b,c} ➞ {True}')
    else:
        print(f'{a,b,c} ➞ {False}')
        
is_triplet(3, 4, 5)
is_triplet(3, 4, 5)
is_triplet(1, 2, 3)
    
```
