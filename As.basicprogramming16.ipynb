```python
1.Write a function that stutters a word as if someone is struggling to read it. The first two letters are repeated twice
with an ellipsis ... and space after each, and then the word is pronounced with a question mark ?
Examples: stutter("incredible") ➞ "in... in... incredible?"
stutter("enthusiastic") ➞ "en... en... enthusiastic?"
stutter("outstanding") ➞ "ou... ou... outstanding?"

Hint :- Assume all input is in lower case and at least two characters long.
```

```python
#1)
def stutter(word):
    y="... "
    z="?"
    try:
        x=((word[0:2]+y+word[0:2]+y+word+z))
        print(f'{word} = {x}')
    except(Exception) as e:
        print("pls enter a string only",e)
        
stutter("incredible")
stutter("enthusiastic")
stutter("outstanding")
```

```python
2..Create a function that takes an angle in radians and returns the corresponding angle in degrees rounded to one decimal place?
Examples: radians_to_degrees(1) ➞ 57.3
radians_to_degrees(20) ➞ 1145.9
radians_to_degrees(50) ➞ 2864.8
```

```python
#2)
import math
def radians_to_degrees(n):
    try:
        x=(n*180)/math.pi
        print(f'{n } radians = {x:.1f}')
    except (Exception) as e:
        print("pls input integers only",e)
radians_to_degrees(1)
radians_to_degrees(20)
radians_to_degrees(50)
```

```python
#3)3.In this challenge, establish if a given integer num is a Curzon number. If 1 plus 2 elevated to num is exactly 
divisible by 1 plus 2 multiplied by num, then num is a Curzon number. Given a non-negative integer num, implement a
function that returns True if num is a Curzon number, or False otherwise.
Examples: `is_curzon(5) ➞ True

# 2 ** 5 + 1 = 33
# 2 * 5 + 1 = 11
# 33 is a multiple of 11
is_curzon(10) ➞ False

# 2 ** 10 + 1 = 1025
# 2 * 10 + 1 = 21
# 1025 is not a multiple of 21
is_curzon(14) ➞ True

# 2 ** 14 + 1 = 16385
# 2 * 14 + 1 = 29
# 16385 is a multiple of 29`
```

```python
#3)
def is_curzon(a):
    try:
        x= (2**a)+1
        y= (2*a)+1
        if x%y== 0:
            print(f'{a} is a curzon number')
        else:
            print(f'{a} is not a curzon number')
    except(Exception) as e:
        print("pls input numbers only",e)
is_curzon(5)
is_curzon(10)
is_curzon(14)
```

```python
#4)4.Given the side length x find the area of a hexagon ?
Examples: area_of_hexagon(1) ➞ 2.6
area_of_hexagon(2) ➞ 10.4
area_of_hexagon(3) ➞ 23.4
```

```python
#4)
import math
def area_of_hexagon(a):
    try:
        hex= (3*math.sqrt(3)/2)*(a**2)
        print(f'area_of_hexagon ({a}) = {hex:.1f}')
    except(Exception) as e:
        print("pls enter integer only",e)
area_of_hexagon(1)
area_of_hexagon(2)
area_of_hexagon(3)
```
