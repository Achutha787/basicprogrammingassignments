```python
#1)1.Create a function that takes a string and returns a string in which each character is repeated once.
Examples:
double_char("String") ➞ "SSttrriinngg"
double_char("Hello World!") ➞ "HHeelllloo WWoorrlldd!!"
doublechar("1234!_") ➞ "11223344!!__"
```

```python
#1)
def double_char(words):
    result= ''
    try:
        for i in words:
            result+= i*2
        return result
    except(Exception) as e:
        print("pls input string only",e)
print(f'➞ {double_char("String")}')
print(f'➞ {double_char("Hello World!")}')
print(f'➞ {double_char("1234!_")}')
    
```

```python
2.Create a function that reverses a boolean value and returns the string "boolean expected" if another variable type
is given.
Examples:
reverse(True) ➞ False
reverse(False) ➞ True
reverse(0) ➞ "boolean expected"
reverse(None) ➞ "boolean expected"
```

```python
#2)
def reverse(boolean):
    if type(boolean) == bool:
        return not boolean
    else:
        return "boolean expected"
print(f'reverse(True) ={reverse(True)}')
print(f'reverse(False)={reverse(False)}')
print(f'reverse(0)= {reverse(0)}')
print(f'reverse(None)={reverse(None)}')
```

```python
3. Create a function that returns the thickness (in meters) of a piece of paper after folding it n number of times.
The paper starts off with a thickness of 0.5mm.
Examples:
`num_layers(1) ➞ "0.001m"

# Paper folded once is 1mm (equal to 0.001m)
num_layers(4) ➞ "0.008m"

# Paper folded 4 times is 8mm (equal to 0.008m)
num_layers(21) ➞ "1048.576m"

# Paper folded 21 times is 1048576mm (equal to 1048.576m)`
```

```python
#3)
def num_layers(num):
    res = 0.5
    for i in range(num):
        res *= 2
    print(f'Output ➞ {res/1000}m') 

num_layers(1)
num_layers(4)
num_layers(21)
```

```python
#4.Create a function that takes a single string as argument and returns an ordered list containing the indices of all
capital letters in the string.
Examples:
index_of_caps("eDaBiT") ➞ [1, 3, 5]
index_of_caps("eQuINoX") ➞ [1, 3, 4, 6]
index_of_caps("determine") ➞ []
index_of_caps("STRIKE") ➞ [0, 1, 2, 3, 4, 5]
index_of_caps("sUn") ➞ [1]
```

```python
def index_of_caps(strings):
    l100=[]
    for i in strings:
        if i.isupper():
            l100.append(strings.index(i))
    print(f'{strings} = {l100}')
index_of_caps("eDaBiT")
index_of_caps("eQuINoX")
index_of_caps("determine")
index_of_caps("STRIKE")
index_of_caps("sUn")
```

```python
5.Using list comprehensions, create a function that finds all even numbers from 1 to the given number.
Examples:
find_even_nums(8) ➞ [2, 4, 6, 8]
find_even_nums(4) ➞ [2, 4]
find_even_nums(2) ➞ [2]
```

```python
#5)
def find_even_nums(n):
    outlist=[i for i in range(1,n+1) if i%2 ==0]
    print(outlist)
find_even_nums(8) 
find_even_nums(4) 
find_even_nums(2)
```
