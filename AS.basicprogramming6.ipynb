```python
#1). Write a Python Program to Display Fibonacci Sequence using Recursion ?
```

```python
#1)
def genFibonacci(n,a,b):
    '''This function will generate fibonacci series of n numbers assked for'''
    try:
        if n == 0:
            return 1
        else:
            result = a+b
            print(result,end=',')
            genFibonacci(n-1,b,result)
    except(Exception) as e:
        print("pls input integers only",e)
in_num = int(input('Enter the length of Series: '))
print('0, 1',end=',')
genFibonacci(in_num,1,2)
```

```python
#2.) Write a Python Program to Find Factorial of a Number using Recursion ?
```

```python
#2)
def factorial(num):
    '''This function will return the factorial of a number'''
    try:
        if num == 0:
            return 1
        else:
            return (num*factorial(num-1))
    except(Exception) as e:
        print("pls input integers only",e)
factorial(6)
```

```python
#3.) Write a Python Program to Calculate your Body Mass Index ?
```

```python
#3)
def calculateBMI():
    try:
        weight = eval(input('Enter your Weight(kgs): '))
        height = eval(input('Enter your Height(mts): '))
        calc_bmi = weight/pow(height,2)
        if (calc_bmi < 18.5):
            status = 'Underweight'
        elif (calc_bmi >= 18.5 and calc_bmi < 24.9):
            status = 'Healthy'
        elif (calc_bmi >= 24.9 and calc_bmi < 30):
            status = 'Overweight'
        elif (calc_bmi >=30):
            status = 'Suffering from Obesity'
        print(f'Your\'re BMI is {calc_bmi} and status is {status} ')
    except(Exception) as e:
        print("pls input positive integers only",e)
calculateBMI()
```

```python
#4.) Write a Python Program to Calculate the Natural Logarithm of any Number ?
```

```python
#4)
import math
def genNatLog():
    '''This function will give lograthamic value of any positive number'''
    try:
        in_num = eval(input("Enter a Number:"))
        print(math.log(in_num))
    except(Exception) as e:
        print("pls input integer only",e)

genNatLog()
```

```python
#5.) Write a Python Program for Cube sum of first n Natural Numbers ?
```

```python
#5)
def cubeOfNaturalNumbers():
    '''this function gives of cube sum of n natural numbers'''
    try:
        num = int(input("Enter the no of Natural Numbers: "))
        result = pow(((num * (num +1))/2),2)
        print(f'The Cube Sum of First {num} Natural Numbers is {result}')
    except(Exception) as e:
        print("pls input natural numbers only",e)

cubeOfNaturalNumbers()
```
