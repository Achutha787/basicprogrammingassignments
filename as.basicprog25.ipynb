```python
def samenumbersum(a,b,c):
    '''this function will return an sum of the output of same numbers.If the user inputs three different inputs,
    it will return zero'''
    try:
        if a == b == c:
            print(f'{a,b,c} ={3}')
        elif a==b or  b==c:
            print(f'{a,b,c}={2}')
        else:
            print(f'{a,b,c} ={0}')
    except(Exception) as  e:
        print("pls input integer only",e)
```

```python

samenumbersum(3, 4, 3)
samenumbersum(1, 1, 1)
samenumbersum(3, 4, 1)
```

```python
#2)
def dictlist(dictin):
    '''this function will return the dictionary into a list of keys-values tuples.'''
    listout = []
    try:
        for keys,values in dictin.items():
            listout.append((keys,values))
        print(f'{dictin} ➞ {listout}')
    except(Exception) as e:
        print("pls input a dictionary",e)
                   
dictlist({"D": 1,"B": 2,"C": 3})
dictlist({"likes": 2,"dislikes": 3,"followers": 10})
```

```python
#3)
def mapping(inlist):
    '''whatever the input list it will covert it into an upper case and makes a dictionary'''
    outlistdict = {}
    try:
        for  i in inlist:
            outlistdict[i]=i.upper()
        print(f'{inlist}➞{outlistdict}')
    except(Exception) as e:
        print("pls enter a list only",e)

mapping(["p", "q"])
mapping(["a", "b", "c"])
mapping(["a", "v", "y", "z"])
```

```python
#4)
def vow_replace(string,chara):
    vowels = ['a','i','o','e','u']
    outresult = ''
    try:
        for j in string:
            if j in vowels:
                outresult += chara
            else:
                outresult += j
        print(f'{string} ➞ {outresult}')
    except(Exception) as e:
        print("pls enter string only",e)
    
vow_replace("apples and bananas",'u')
vow_replace("cheese casserole", "o")
vow_replace("stuffed jalapeno poppers", "e")
```

```python
#5)
def ascii_capitalize(instring1):
    ''' function that takes a string as input and capitalizes a letter if its ASCII code is even
and returns its lower case version if its ASCII code is odd.'''
    outstring1 = ''
    try:
        for i in instring1.lower():
            if (ord(i)%2 == 0):
                outstring1 += i.upper()
            else:
                outstring1 += i
        print(f'{instring1} = {outstring1}')
    except(Exception) as e:
        print("pls input string only",e)
        
ascii_capitalize("to be or not to be!")
ascii_capitalize("THE LITTLE MERMAID")
ascii_capitalize("Oh what a beautiful morning.")
```
