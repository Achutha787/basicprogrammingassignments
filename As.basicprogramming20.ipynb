```python
1.Create a function that takes a list of strings and integers,and filters out the list so that it returns a list of integer
s only.
Examples:
filter_list([1, 2, 3, "a", "b", 4]) ➞ [1, 2, 3, 4]
filter_list(["A", 0, "Edabit", 1729, "Python", "1729"]) ➞ [0, 1729]
filter_list(["Nothing", "here"]) ➞ []
```

```python
def filter_list(l1):
    l2=[]
    try:
        for i in l1:
            if type(i) == int:
                l2.append(i)
        print(f'{l1} = {l2}')
    except(Exception) as e:
        print("take list only",e)

filter_list([1, 2, 3, "a", "b", 4]) 
filter_list(["A", 0, "Edabit", 1729, "Python", "1729"])
filter_list(["Nothing", "here"]) 
```

```python
2.Given a list of numbers,create a function which returns the list but with each element's index in the list added to itself
This means you add 0 to the number at index 0, add 1 to the number at index 1, etc...
Examples:
add_indexes([0, 0, 0, 0, 0]) ➞ [0, 1, 2, 3, 4]
add_indexes([1, 2, 3, 4, 5]) ➞ [1, 3, 5, 7, 9]
add_indexes([5, 4, 3, 2, 1]) ➞ [5, 5, 5, 5, 5]
```

```python
#2)
def add_indexes(l5):
    l6=[]
    sum = 0
    try:
        for i in l5:
            sum = i+l5.index(i)
            l6.append(sum)
        print(l6)
    except(Exception) as e:
        print("pls input integers of list only",e)
        
add_indexes([0, 0, 0, 0, 0]) 
add_indexes([1, 2, 3, 4, 5]) 
add_indexes([5, 4, 3, 2, 1]) 
```

```python
#3)
3.Create a function that takes the height and radius of a cone as arguments and returns the volume of the cone rounded 
to the nearest hundredth. See the resources tab for the formula.
Examples:
cone_volume(3, 2) ➞ 12.57
cone_volume(15, 6) ➞ 565.49
cone_volume(18, 0) ➞ 0
```

```python
#3)
import math
def cone_volume(a,b):
    output = ((math.pi)*pow(b,2))*(a/3)
    print(f'Output ➞ {output:.2f}')

cone_volume(3,2) 
cone_volume(15,6)    
cone_volume(18,0)    
```

```python
4.This Triangular Number Sequence is generated from a pattern of dots that form a triangle.
The first 5 numbers of the sequence, or dots, are: 1, 3, 6, 10, 15
This means that the first triangle has just one dot, the second one has three dots, the third one has 6 dots and so on.
Write a function that gives the number of dots with its corresponding triangle number of the sequence.

Examples:
triangle(1) ➞ 1
triangle(6) ➞ 21
triangle(215) ➞ 23220
```

```python
#4)
def triangle(num):
    print(f'Output ➞ {int((num)*((num+1)/2))}')

triangle(1)
triangle(6)
triangle(215)
```

```python
#5)
def missing_num(lst):
    try:
        for i in range(1,11):
            if i not in lst:
                print(f'{lst} ={i}')
    except(Exception) as e:
        print("input list of numbers only",e)
        
missing_num([1, 2, 3, 4, 6, 7, 8, 9, 10]) 
missing_num([7, 2, 3, 6, 5, 9, 1, 4, 8]) 
missing_num([10, 5, 1, 2, 4, 6, 8, 3, 9]) 
```
