```python
#1)
def sumOfArray():
    in_array = eval(input("Enter the Array: "))
    print(f'The sum of {in_array} is {sum(in_array)}')

sumOfArray()
```

```python
#2)
def largestelementarray():
    in_array= eval(input("Enter the array: "))
    print(f'largestelement of array is{in_array}:{sorted(in_array,reverse=True)[0]}')
largestelementarray()        
```

```python
#3) Write a Python Program for array rotation ?
```

```python
#3)
def arrayrotation():
    in_array = eval(input("enter the array"))
    print(f'arrayrotation of {in_array} : {in_array[::-1]}')
arrayrotation()
```

```python
#5) Write a Python Program to check if given array is Monotonic ?
```

```python
#5)
def checkMonotonic():
    in_arr = eval(input("Enter the Array: "))
    if(all(in_arr[i]<=in_arr[i+1]) for i in range(len(in_arr)-1)) or all(in_arr[i]>=in_arr[i+1] for i in range(len(in_arr)-1))):
        print(f'Array {in_arr} is Monotonic')
    else:
        print(f'Array {in_arr} is Not Monotonic')

checkMonotonic()
checkMonotonic()
```
